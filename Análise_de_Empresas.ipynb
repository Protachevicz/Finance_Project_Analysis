# **Listando as Ações segundo a Fórmula Mágica**:

Joel Greenblatt, que é professor de finanças e gestor de fundos de investimentos, fez um livro para essa série, chamado “The Little Book That Beats the Market” (“O pequeno livro que vence o mercado”, em tradução livre e literal), em 2005.

O livro foi traduzido para o Português e publicado no Brasil com o nome “O Mercado de Ações ao Seu Alcance” e, posteriormente, em uma nova edição, como “A Fórmula Mágica de Joel Greenblatt para Bater o Mercado de Ações“.

---
---



A ideia de Joel foi criar uma forma de seleção de ativos que fosse automatizada, através de métodos quantitativos, e para isso definiu dois filtros:

1. Selecionar empresas de qualidade através do indicador **ROIC**(retorno sob o capital investido).
2. Selecionar ações com preços descontados utilizando o **EV/EBIT** (relação do valor de mercado da firma com o seu lucro operacional).

A Magic Formula é uma maneira simples de encontrar firmas que possuem grandes margens de lucro nas suas operações, mas que estão sendo negociadas a um baixo preço na bolsa. Através da fórmula cria-se um ranking de empresas que oferecem as melhores condições para investimento no momento.

Sendo assim temos duas métricas principais: uma para mostrar empresas que são “bem administradas”, ou seja, aquelas que oferecem bons retornos em suas operações, e outra para mostrar empresas que estão sendo negociadas com “desconto”, a um preço menor do que ela “realmente vale”.

**Como montar a tabela usando a fóruma mágica**: Com os dois indicadores, deve-se construir um ranking delas. Se faz, então, uma lista para os dois indicadores, atribuindo uma nota crescente conforme o valor do indicador. Assim, a empresa com o maior Earnings Yield (EV/EBIT) recebe a nota 1, a segunda recebe nota 2 e assim por diante.

A mesma coisa é feita para o RIOC: Se faz uma lista (com as mesmas empresas) onde aquela com ROIC maior recebe nota 1 e assim sucessivamente.

Aí, se faz um terceiro ranking, com as notas SOMADAS dos dois indicadores, dispondo aquelas com MENORES notas na frente (pois representam as “melhores” empresas pela combinação dos dois indicadores).
Assim, neste terceiro e último ranking (que é o ranking combinado dos dois indicadores), uma nota baixa representa uma empresa “melhor”. Por isso, elas devem ser dispostas pela nota em ordem crescente.

**EV/EBIT** é um indicador financeiro que compara o Valor da Firma (EV ou Enterprise Value) com o Lucro Antes de Impostos e Taxas, o EBIT. A principal função do EV/EBIT é ajudar a identificar quanto uma empresa custa em relação ao que ela produz a partir de sua atividade fim.

Quando o EV/EBIT está elevado, existe uma indicação de que a empresa possui uma boa avaliação no mercado. Consequentemente, isso significa que suas ações estão valorizadas.

Enquanto isso, um EV/EBIT baixo demonstra que a empresa está sendo subavaliada. Com isso, possuem uma forte tendência de valorização ao longo do tempo, tornando a ação mais atrativa para a compra.



---


**ROIC** O Return on Invested Capital (ROIC), também conhecido como retorno sobre capital investido, é um indicador financeiro que aponta o valor que uma empresa gera com base no valor investido. Ou seja, esta métrica serve para apontar se os investimentos realizados por um negócio surtiram efeitos positivos ou não.

Assim, o retorno sobre capital investidor se torna um dado utilizado por investidores que desejam avaliar a capacidade que uma empresa possui de gerar riqueza. Afinal, este dado mostra qual a rentabilidade que uma empresa possui frente ao valor investido por ela.

Inicialmente, é importante reforçar que este indicador financeiro considera a capacidade de retorno que uma empresa tem através do valor que investiu em si mesma. Dessa forma, diversos investidores consideram que: ROIC, quanto maior melhor.

Em suma, esta ideia é baseada na ideia que um retorno sobre capital investido elevado aponta um bom desempenho financeiro e administrativo por parte da empresa.

Enquanto, quando o ROIC apresentado for baixo, este pode ser um indicativo que tanto gestão quanto finanças do negócio não estão bem.
